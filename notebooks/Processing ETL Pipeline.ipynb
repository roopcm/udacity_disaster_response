# Processing ETL Pipeline
Instructions to create an ETL
### A. Import libraries and datasets.
- Import Python libraries
- Load `messages.csv` into a dataframe and inspect the first few lines.
- Load `categories.csv` into a dataframe and inspect the first few lines.

```python
# import libraries
import pandas as pd
from sqlalchemy import create_engine
from langdetect import detect
```

```python
# load messages dataset
messages = pd.read_csv('../data/disaster_messages.csv')
messages.head()
```

```python
# load categories dataset
categories = pd.read_csv('../data/disaster_categories.csv')
categories.head()
```

### B. Datasets Merging.
- Merge the messages and categories datasets using the key 'id'
- Name the combined dataset df which we will clean as we move along 

```python
# merge datasets
df = messages.merge(categories, how='outer',on='id')
df.head()
```

### C. Split `categories` into separate category columns.
- Split the values in the `categories` column on the `;`  so that each value splits to  a separate column. 
- Rename columns of `categories` with new column names.

```python
# create a dataframe of the 36 individual category columns
categories = df['categories'].str.split(';', expand = True)
categories.head()
```

```python
# select the first row of the categories dataframe which we use to extract the list of new column names.(looking 
### at the pattern for each case we need to extract only the characters except the last 2 ) 
row = categories.iloc[0,:]
category_colnames = row.transform(lambda x: x[:-2]).tolist()
print(category_colnames)
```

```python
# rename the columns of `categories`
categories.columns = category_colnames
categories.head()
```

### D. Quantification of category Values 
Convert category values to just numbers 0 or 1.
- Iterate through the category columns in df to keep only the last character of each string (the 1 or 0). 

```python
for column in categories:
    # set each value to be the last character of the string
    categories[column] = categories[column].transform(lambda x: x[-1:])
    
    # convert column from string to numeric
    categories[column] = categories[column].astype(int)
categories.head()
```

### E. Replace `categories` column in `df` with new category columns.
- Drop the categories column from the df dataframe since it is no longer needed.
- Concatenate df and categories data frames.

```python
# drop the original categories column from `df`
df = df.drop('categories',axis=1)

df.head()
```

```python
# concatenate the original dataframe with the`categories` dataframe
df = pd.concat([df, categories], axis = 1)
df.head()
```

```python
#We check here the number of target categories
for col in df.columns:
    number_of_unique = len(df[col].unique())
    if number_of_unique != 2:
        print(col,number_of_unique)
```
